```python
'''
  * Else
  * 
'''

tinggiBadan = 120

if(tinggiBadan >= 160): #True / Bener
  print("Tinggi Sekalih")
else: # False / salah
  print("Kosong")
```

```python
'''
  * Else If
  * else if dengan dua kondisi bener / true
'''

name = "Nelan"
score = 90

if(score >= 90):
  print(f"Selamat {name} mendapatkan grade A")
elif(score >= 80):
  print(f"Selamat {name} mendapatkan grade B")
elif(score >= 70):
  print(f"Selamat {name} mendapatkan grade C")
elif(score >= 60):
  print(f"Selamat {name} mendapatkan grade D")
else:
  print(f"Selamat {name} mendapatkan grade E")


```
