```python
'''
  * For Loop
'''

manchesterUnited = ["Ronaldo", "Rooney", "Rafael", "Nani", "Anderson"]



i = 0
for player in manchesterUnited:
  print(i, player)
  i += 1
```

```python
# Animals

animals = ["Ikan", "Kucing", "Buaya", "Kuda"]


print(animals[0])
print(animals[1])
print(animals[2])
print(animals[3])
print(animals[4])

for animal in animals:
  print(animal)
```
