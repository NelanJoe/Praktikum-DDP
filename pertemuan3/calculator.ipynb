```python
'''
 * Calculator sederhana
'''

bilanganPertama = int(input("Masukkan bilangan pertama: "))
bilanganKedua = int(input("Masukkan bilangan kedua: "))

operator = input("Masukkan jenis operator: ")

# Pengkondisian jenis operator
total = 0
if(operator == "+"):
  total = bilanganPertama + bilanganKedua
  print(total)
elif(operator == "-"):
  total = bilanganPertama - bilanganKedua
  print(total)
elif(operator == "*"):
  total = bilanganPertama * bilanganKedua
  print(total)
elif(operator == "/"):
  total = bilanganPertama / bilanganKedua
  print(total)
elif(operator == "%"):
  total = int(bilanganPertama % bilanganKedua)
  print(total)
else:
  print("Operator yang kamu masukkan salah")
```
